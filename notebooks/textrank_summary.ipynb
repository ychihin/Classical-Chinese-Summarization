```python
# Pick top k sentence from sen_num sentences text
sen_num = 20
k = 3
```

```python
# Get first sen_num number sentences for the sample file
text = []
with open('../data/processed/sample/tran_en.txt', 'r') as file:
    for i in range(sen_num):
        line = file.readline()
        text.append(line)
        if not line:
            break

for sen in text:
    print(sen)
```

```python
word_to_id = dict()
```

```python
from collections import Counter
import math

# Compute word tf-idf score
def get_tfidf(text):
    tf_idf = []
    sen_word_bag_list = []
    tf_list = []
    for sen in text:
        sen = sen.lower().split()
        sen_word_count = Counter(sen)
        sen_word_bag = set(sen)
        sen_word_bag_list.append(Counter(sen_word_bag))
        sen_tf = [(word,sen_word_count[word]/len(sen)) for word in sen_word_count]
        tf_list.append(sen_tf)
    word_count = sum(sen_word_bag_list, Counter())
    idf = {word:math.log(sen_num/count) for word,count in word_count.items()}
    tf_idf = [[(word[0],word[1]*idf[word[0]]) for word in tf] for tf in tf_list]
    return tf_idf, word_count
```

```python
tf_idf, word_count = get_tfidf(text)
tf_idf
```

```python
# Calculate per sentence tf-idf score
sen_score = [sum([word_tf_idf[1] for word_tf_idf in sen_tf_idf]) for sen_tf_idf in tf_idf]
print(sen_score)
```

```python
# Create a word to id mapping
word_to_id = {word:idx for idx,word in enumerate(word_count.keys())}
id_to_word = {id:word for word,id in word_to_id.items()}
print(word_to_id)
print(id_to_word)
```

```python
# Show top k sentence indices
indices = [i for i, v in sorted(enumerate(sen_score), key=lambda x: x[1], reverse=True)[:k]]
indices
```

```python
# Print summarization result based on top k sentence with sum tf-idf score
for i in sorted(indices):
    print(text[i])
```

```python
import numpy as np

# Create and populate a word score sparse matrix
tfidf_matrix = np.zeros((sen_num, len(word_to_id)))
for i,sen in enumerate(tf_idf):
    for word in sen:
        score = word[1]
        tfidf_matrix[i][word_to_id[word[0]]] = score
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Create edge score matrix and normalize
edge_score = cosine_similarity(tfidf_matrix)
np.fill_diagonal(edge_score, 0)
row_sums = edge_score.sum(axis=1, keepdims=True)
edge_score = edge_score / row_sums
edge_score
```

```python
# Iterate to calcualte textrank
d = 0.85
max_iter = 1000
convergence = 1e-6
scores = [1.0] * sen_num 

iter_count = 0
for _ in range(max_iter):
    new_scores = [(1 - d) for _ in range(sen_num)]

    for i in range(sen_num):
        for j in range(sen_num):
            if i != j:
                new_scores[i] += d * edge_score[j][i] * scores[j]
    diff = sum(abs(new_scores[i] - scores[i]) for i in range(sen_num))
    if diff < convergence:
        break
    scores = new_scores
    iter_count += 1
print(iter_count)
scores
```

```python
# Show top k sentence indices
indices = [i for i, v in sorted(enumerate(scores), key=lambda x: x[1], reverse=True)[:k]]
indices
```

```python
# Print summarization result based on textrank
for i in sorted(indices):
    print(text[i])
```
