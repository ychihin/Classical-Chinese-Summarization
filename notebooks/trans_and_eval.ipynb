```python
from googletrans import Translator

translator = Translator()
```

```python
# Present first 5 results
with open('../data/raw/sample/target.txt', 'r') as file:
    for i,line in enumerate(file):
        if i > 5:
            break
        result = await translator.translate(line, src='zh-cn', dest='en')
        print(result.text)
        print(line)
```

```python
# Translate to English and save all results to a file
with open('../data/processed/sample/tran_en.txt', 'w') as out:
    with open('../data/raw/sample/target.txt', 'r') as file:
        for line in file:
            result = await translator.translate(line, src='zh-cn', dest='en')
            out.write(result.text)
            out.write('\n')
```

```python
# Translate back to Chinese and save all results to a file
with open('../data/processed/sample/tran_cn.txt', 'w') as out:
    with open('../data/processed/sample/tran_en.txt', 'r') as file:
        for line in file:
            result = await translator.translate(line, src='en', dest='zh-cn')
            out.write(result.text)
            out.write('\n')
```

```python
# Present source and target line by line
with open('../data/processed/sample/tran_cn.txt', 'r') as target, open ('../data/raw/sample/target.txt', 'r') as source:
    for s, t in zip(source, target):
        print(f'Source: {s}Target: {t}')
```
